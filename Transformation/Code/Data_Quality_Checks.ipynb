```python
#● Null Checks: Identify and handle null values in essential fields
def Nulls_Checks(df):
    return df.isnull().sum()/len(df)*100



def Replace_nulls(df,col):

    if df[col].dtype == 'str' :
      df[col]=df[col].fillna('Unknown')

    elif df[col].dtype in ['int64','float64']:
        df[col]=df[col].fillna(0)

    elif df[col].dtype == 'datetime64[ns]':
        df[col]=df[col].fillna(pd.Timestamp.today())

    return df 
```

```python
Nulls_Checks(df)
```

```python
Replace_nulls(df,'phone')
```

```python
Nulls_Checks(df)
```

```python
#● Duplicate Checks: Detect and remove duplicate rows.
def duplicate_checks(df):
    total           = len(df)
    duplicate_count = df.duplicated().sum()
    duplicate_pct   = duplicate_count / total * 100

    print(f"Total Rows       : {total}")
    print(f"Duplicate Rows   : {duplicate_count}")
    print(f"Duplicate %      : {duplicate_pct.round(2)}%")

    if duplicate_count > 0:
        return df[df.duplicated(keep=False)]  # return the duplicate rows
    else:
        print("No duplicates found!")
        return None

duplicate_checks(df)
```

```python
# Maybe a customer_id appears twice — even if other columns differ
def duplicate_checks_Column(df, col=None):
    total = len(df)

    if col:
        duplicate_count = df.duplicated(subset=[col]).sum()
        print(f"Duplicates in '{col}': {duplicate_count} ({duplicate_count/total*100:.2f}%)")
    else:
        duplicate_count = df.duplicated().sum()
        print(f"Full Row Duplicates : {duplicate_count} ({duplicate_count/total*100:.2f}%)")

    if duplicate_count > 0:
        return df[df.duplicated(subset=[col] if col else None, keep=False)]
    else:
        print("No duplicates found!")
        return None
```

```python
duplicate_checks_Column(df,'order_id')
#its normal to have duplicates in order_id 
```

```python
#some validations 
# الخصم المفروض يكون بين 0 و 1 بس (0% → 100%)
def validate_discount(df):
    invalid = df[(df['discount'] < 0) | (df['discount'] > 1)]
    print(f"Invalid discounts: {len(invalid)} rows")
    return invalid

validate_discount(df)
```
